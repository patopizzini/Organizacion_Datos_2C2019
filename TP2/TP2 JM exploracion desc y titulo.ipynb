```python
#Importacion de librerias
import pandas as pd
import numpy as np
import datetime as DT
import warnings

from collections import Counter
from sklearn.preprocessing import LabelEncoder

warnings.filterwarnings('ignore')
```

```python
#Lectura de archivos CSV
train = pd.read_csv('../../DATA_TP2/train.csv')
test = pd.read_csv('../../DATA_TP2/test.csv')
```

```python
#Nos deshacemos de los NaN
train['titulo'] = train['titulo'].dropna()
train['descripcion'] = train['descripcion'].dropna()

#Uppercase
train['titulo'] = train['titulo'].str.upper()
train['descripcion'] = train['descripcion'].str.upper()

#Limpiamos las columnas
train['titulo'] = train['titulo'].str.normalize('NFKD')\
                                       .str.encode('ascii', errors='ignore')\
                                       .str.decode('utf-8')
train['descripcion'] = train['descripcion'].str.normalize('NFKD')\
                                       .str.encode('ascii', errors='ignore')\
                                       .str.decode('utf-8')

#Separamos en palabras las columnas que necesitamos parsear
train['titulo_palabras'] = train['titulo'].str.strip().str.split('[\W_]+')
train['descripcion_palabras'] = train['titulo'].str.strip().str.split('[\W_]+')
```

```python
#Armamos una lista con las palabras que mas se repiten en la columna titulo
#Decartamos todos los resultados con longitud menor a 3
prueba = train['titulo'].tolist()
prueba2 = list()

for i in prueba:
    for j in str(i).split():
        if len(j) > 3:
            prueba2.append(j)

contados = Counter(prueba2)
contados.most_common()
```
